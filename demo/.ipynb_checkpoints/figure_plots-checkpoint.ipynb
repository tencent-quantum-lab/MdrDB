```python
import pandas as pd
import numpy as np

import matplotlib as mpl
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./1-MdrDB_full_features1.csv", sep=",")
# print methods
print([i for i in data[data["Scenario"] == "Scenario1"]["Methods"]])
```

```python
method = ['AdaBoost', 'ElasticNet', 'Lasso', 'MLP', 'DecisionTree', 'SVR', 'RandomForest', 'Bagging', 'GradientBoost', 'ExtraTrees']
data.head()
```

```python
critiers = ["RMSE", "Pearson", "AUPRC"]
x = np.arange(len(list(data["Methods"].unique())))  # x轴刻度标签位置
labels = list(data["Methods"].unique())
scenario = ["Scenario1", "Scenario2", "Scenario3"]
width = 0.25  # the width of the bars

fig, axes = plt.subplots(figsize=(15, 7), nrows=3, ncols=1)
for i, critier in enumerate(critiers):
    axes[i].bar(x - width, data[data["Scenario"] == "Scenario1"][critier], width, label="Platinum (no tyrosine kinase)", color="#ffd166")
    axes[i].bar(x, data[data["Scenario"] == "Scenario2"][critier], width, label="Platinum", color="#ed6a5a")
    axes[i].bar(x + width, data[data["Scenario"] == "Scenario3"][critier], width, label="MdrDBCoreset", color="#20a39e")
    axes[i].set_ylabel(critier, fontsize=15)    
    axes[i].set_xticks(x, labels=method, rotation=0, fontsize=13)

plt.legend(bbox_to_anchor=(0.72, -0.2), ncol=3, fontsize=15)
plt.tight_layout()
plt.savefig("./Figure1_MdrDB_coreset_TKI_barplot.pdf", dpi=300, bbox_inches='tight')
plt.show()
```
